```python
import glob
import torch
import torchvision
import matplotlib.pyplot as plt

from nice import Nice

DEVICE = "cuda"
```

```python
model = Nice(784, device=DEVICE).to(DEVICE)
model = model.to(DEVICE)
model.eval()
```

```python
# Used when training the model and comparing different checkpoints

def load_model_file(comparater_fn=min, comparater_key_field="epoch"):
    files = sorted(glob.glob("./nice_epoch*.pt"))
    
    files_data = []
    for file in files:
        f_break = file.split("_")
        epoch = f_break[f_break.index("epoch") + 1]
        train_loss = f_break[f_break.index("epoch") + 4]
        val_loss = f_break[f_break.index("val") + 2].split(".")[0]
        files_data.append({
            "filename": file,
            "epoch": int(epoch),
            "train_loss": float(train_loss),
            "val_loss": float(val_loss)
        })

    best = comparater_fn(files_data, key=lambda d : d[comparater_key_field])
    
    return best["filename"]
```

```python
def display_samples(is_training=False):
    # When is_training is True, we will load the best checkpoint based on
    # latest epoch save, first 3x3 samples, and lowest validation loss, last
    # 3x3 samples.

    rows, cols = 3, 3
    t = rows*cols
    _, ax = plt.subplots(2*rows, cols)
    ax = ax.flatten()

    for k, (comparater_fn, comparater_key_field) in enumerate([(max, "epoch"), (min, "val_loss")]):
        if is_training:
            model_save_file = load_model_file(comparater_fn, comparater_key_field)
            print("Using file:", model_save_file)
        else:
            model_save_file = "model.pt"

        model.load_state_dict(torch.load(model_save_file))
        for i in range(t):
            x_hat = model.sample(n=1).reshape(28,28)
            img = torchvision.transforms.ToPILImage()(x_hat)

            ax[k*t + i].imshow(img)
            ax[k*t + i].axis("off")

display_samples()
```

```python
def ema(signal, alpha):
    ema = torch.zeros_like(signal)
    ema[0] = signal[0]
    for t in range(1, len(signal)):
        ema[t] = alpha * signal[t] + (1 - alpha) * ema[t - 1]
    return ema

loss_curves = torch.load("loss_curves.pt")
loss_curves = ema(loss_curves, 1)

fig = plt.figure()

plt.plot(loss_curves[0].clamp(max=1500).cpu().detach().numpy())
plt.plot(loss_curves[1].clamp(max=1500).cpu().detach().numpy())
plt.legend(["Train", "Validation"])

plt.show()
```
